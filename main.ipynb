### Import Packages

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Vanilla Image Segmentation

### Load Images

```python
image = cv2.imread('1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image)
plt.axis('off')
plt.show()
```

### Convert Images into Greyscales

```python
# Convert to grayscale
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Display the image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

### Apply global thresholding where threshold = 110

```python
# Apply global thresholding with threshold = 110
_ , image = cv2.threshold(image, 110, 255, cv2.THRESH_BINARY)

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

### Apply Area Based Image Segmentation

```python

num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(image)
min_area_threshold = 200
mask = np.zeros_like(image)

for label in range(1, num_labels):
    area = stats[label, cv2.CC_STAT_AREA]
    if area > min_area_threshold:
        mask[labels == label] = 255

# Display the image
plt.imshow(mask, cmap='gray')
plt.axis('off')
plt.show()
```

### Honorable Mention: Trial using morphology opening and closing (dropped)

```python
kernel = (3,3)
kernel_close = (4,4)

# Perform opening and closing operation
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, cv2.getStructuringElement(cv2.MORPH_RECT, kernel))
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_RECT, kernel_close))

mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, cv2.getStructuringElement(cv2.MORPH_RECT, kernel))
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_RECT, kernel_close))

mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, cv2.getStructuringElement(cv2.MORPH_RECT, kernel))
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_RECT, kernel_close))

mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, cv2.getStructuringElement(cv2.MORPH_RECT, kernel))
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_RECT, kernel_close))

# Perform closing operation twice
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_RECT, kernel_close))
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_RECT, kernel_close))

# Display the image
plt.imshow(mask, cmap='gray')
plt.axis('off')
plt.show()
```
